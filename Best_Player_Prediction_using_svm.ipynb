```python
def my_func(my_df,team_first,team_2nd):
    batsmen=[]
    score=[]
    balls=[]
    for i,v in my_data.iterrows():
        if v.batting_first_second==team_first:
            batsmen.append(v.Batsman)
            score.append(v["Runs-off-bat"])
            balls.append(v.Overs)

        
    batsmen_dataframe=pd.DataFrame(batsmen,columns=["Batsman"])
    score_dataframe=pd.DataFrame(score,columns=["Runs-off-bat"])
    balls_dataframe=pd.DataFrame(balls,columns=["Overs"])
    batsmen_score=pd.concat([batsmen_dataframe, score_dataframe,balls_dataframe],axis=1)

    player_score=batsmen_score.groupby(["Batsman"]).agg({"Runs-off-bat":["sum"]})
    Team_score=player_score['Runs-off-bat'].sum()
    
    player_score["player score / Team score"]=player_score["Runs-off-bat"]/Team_score
    
    batsmen_ball=batsmen_score.groupby(['Batsman']).count()
    batsmen_ball=batsmen_ball.reset_index()
    batsmen_ball=batsmen_ball.drop(columns=['Runs-off-bat'],axis=1)
    batsmen_ball.columns=['Batsmen','Balls']
    Team_balls=batsmen_ball["Balls"].sum()
    
    batsmen_ball["Ball_batsmen_played / Team balls"]=batsmen_ball.Balls/Team_balls
    
    
    O_t_score=[]
    O_t_batsmen=[]
    for j,k in my_data.iterrows():
        if k.batting_first_second==team_2nd:
            O_t_score.append(k["Runs-off-bat"])
            O_t_batsmen.append(k.Batsman)
        
        
    O_t_batsmen_dataframe=pd.DataFrame(O_t_batsmen,columns=["Batsmen"])        
    O_t_score_dataframe=pd.DataFrame(O_t_score,columns=["Runs-off-bat"])
    O_t_batsmen_score=pd.concat([O_t_batsmen_dataframe, O_t_score_dataframe],axis=1)

    O_t_player_score=O_t_batsmen_score.groupby(["Batsmen"]).agg({"Runs-off-bat":["sum"]})
    O_t_total_score=O_t_player_score['Runs-off-bat'].sum()
    
    player_score["player score / O_T_S"]=player_score["Runs-off-bat"]/O_t_total_score
    
    player_score=player_score.reset_index()
    
    player_score["Ball_batsmen_played / Team balls"]=batsmen_ball["Ball_batsmen_played / Team balls"]
    
    
    player_score["wining player"]=player_score['player score / Team score']+player_score['player score / O_T_S']+player_score['Ball_batsmen_played / Team balls']
    
    wining=[]
    for i in player_score['wining player']:
        if i<0.45:
            i=0
            wining.append(i)
        else:
            i=1
            wining.append(i)
    player_score["wining player"]=wining
    final_label_24=pd.DataFrame(player_score)
    
    
    
    
    for_2nd_team_batsmen=[]
    for_2nd_team_score=[]
    for_2nd_team_balls=[]
    for i,v in my_data.iterrows():
        if v.batting_first_second==team_2nd:
            for_2nd_team_batsmen.append(v.Batsman)
            for_2nd_team_score.append(v["Runs-off-bat"])
            for_2nd_team_balls.append(v.Overs)

        
    for_2nd_team_batsmen_dataframe=pd.DataFrame(for_2nd_team_batsmen,columns=["Batsman"])
    for_2nd_team_score_dataframe=pd.DataFrame(for_2nd_team_score,columns=["Runs-off-bat"])
    for_2nd_team_balls_dataframe=pd.DataFrame(for_2nd_team_balls,columns=["Overs"])
    for_2nd_team_batsmen_score=pd.concat([for_2nd_team_batsmen_dataframe, for_2nd_team_score_dataframe,for_2nd_team_balls_dataframe],axis=1)

    
    for_2nd_team_player_score=for_2nd_team_batsmen_score.groupby(["Batsman"]).agg({"Runs-off-bat":["sum"]})
    for_2nd_team_Team_score=for_2nd_team_player_score['Runs-off-bat'].sum()

    
    for_2nd_team_player_score["player score / Team score"]=for_2nd_team_player_score["Runs-off-bat"]/for_2nd_team_Team_score


    
    for_2nd_team_batsmen_ball=for_2nd_team_batsmen_score.groupby(['Batsman']).count()
    for_2nd_team_batsmen_ball=for_2nd_team_batsmen_ball.reset_index()
    for_2nd_team_batsmen_ball=for_2nd_team_batsmen_ball.drop(columns=['Runs-off-bat'],axis=1)
    for_2nd_team_batsmen_ball.columns=['Batsmen','Balls']
    for_2nd_team_Team_balls=for_2nd_team_batsmen_ball["Balls"].sum()
    
    
    for_2nd_team_batsmen_ball["Ball_batsmen_played / Team balls"]=for_2nd_team_batsmen_ball.Balls/for_2nd_team_Team_balls


    for_2nd_team_O_t_score=[]
    for_2nd_team_O_t_batsmen=[]
    for j,k in my_data.iterrows():
        if k.batting_first_second==team_first:
            for_2nd_team_O_t_score.append(k["Runs-off-bat"])
            for_2nd_team_O_t_batsmen.append(k.Batsman)
            
    for_2nd_team_O_t_batsmen_dataframe=pd.DataFrame(for_2nd_team_O_t_batsmen,columns=["Batsmen"])        
    for_2nd_team_O_t_score_dataframe=pd.DataFrame(for_2nd_team_O_t_score,columns=["Runs-off-bat"])
    for_2nd_team_O_t_batsmen_score=pd.concat([for_2nd_team_O_t_batsmen_dataframe, for_2nd_team_O_t_score_dataframe],axis=1)

    
    for_2nd_team_O_t_player_score=for_2nd_team_O_t_batsmen_score.groupby(["Batsmen"]).agg({"Runs-off-bat":["sum"]})
    for_2nd_team_O_t_total_score=for_2nd_team_O_t_player_score['Runs-off-bat'].sum()

    
    for_2nd_team_player_score["player score / O_T_S"]=for_2nd_team_player_score["Runs-off-bat"]/for_2nd_team_O_t_total_score

    
    for_2nd_team_player_score=for_2nd_team_player_score.reset_index()
    
    for_2nd_team_player_score["Ball_batsmen_played / Team balls"]=for_2nd_team_batsmen_ball["Ball_batsmen_played / Team balls"]
    
    
    for_2nd_team_player_score["wining player"]=for_2nd_team_player_score['player score / Team score']+for_2nd_team_player_score['player score / O_T_S']+for_2nd_team_player_score['Ball_batsmen_played / Team balls']

    
    for_2nd_team_wining=[]
    for i in for_2nd_team_player_score['wining player']:
        if i<0.45:
            i=0
            for_2nd_team_wining.append(i)
        else:
            i=1
            for_2nd_team_wining.append(i)
    for_2nd_team_player_score["wining player"]=for_2nd_team_wining
    for_2nd_team_final_label_24=pd.DataFrame(for_2nd_team_player_score)


    
    
    return final_label_24,for_2nd_team_final_label_24
```
